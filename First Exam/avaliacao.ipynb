Agora vamos utilizar a biblioteca Matplotlib para criar um gráfico de barras que exibe a contagem de gêneros no DataFrame "df".

```python
#importação de cores para estilizar o gráfico

cores = ["#89CFF0", "#FF69B4", "#FFD700", "#7B68EE", "#FF4500",
          "#9370DB", "#32CD32", "#8A2BE2", "#FF6347", "#20B2AA",
          "#FF69B4", "#00CED1", "#FF7F50", "#7FFF00", "#DA70D6"]
```

```python
# Define o tamanho da figura
plt.figure(figsize=(20, 6))

# Cria um gráfico de barras com contagem de gêneros no DataFrame "df"
ax = df["Gender"].value_counts().plot(kind = 'bar', color = cores, rot = 0)

# Define os rótulos do eixo x como "Male" e "Female"
ax.set_xticklabels(('Male', 'Female'))

# Adiciona rótulos com o valor de cada barra acima das barras
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha='center', va='bottom', color='black')
    
# Define o tamanho das fontes dos rótulos dos eixos
ax.tick_params(axis='both', labelsize=15)

# Define o rótulo do eixo x
plt.xlabel('Gênero', weight="bold", color="black", fontsize=14, labelpad=20)

# Define o rótulo do eixo y
plt.ylabel('Número de Ocorrências', weight="bold", color="black", fontsize=14, labelpad=20)

# Exibe o gráfico
plt.show()
```

O gráfico a seguir combina um histograma para mostrar a distribuição discreta das idades em faixas específicas e uma curva de densidade para mostrar como as idades se distribuem de forma contínua em todo o espectro. Ele fornece uma representação mais completa da distribuição das idades no conjunto de dados.

```python
# Cria uma figura e um eixo (subplot) com o tamanho especificado
fig, ax = plt.subplots(figsize=(20, 5))

# Plota um histograma da coluna 'Age' do DataFrame 'df'
ax.hist(df['Age'], bins=25, edgecolor='black', alpha=0.7, color='skyblue', density=True)

# Plota uma curva de densidade (KDE) da coluna 'Age' em vermelho no mesmo eixo
df['Age'].plot(kind='kde', color='red', ax=ax)

# Define rótulos para os eixos x e y e um título para o gráfico
ax.set_xlabel('Idade')
ax.set_ylabel('Contagem / Densidade')
ax.set_title('Distribuição de Idade - Histograma com Curva de Densidade')

# Adiciona uma legenda para o histograma e a curva de densidade
ax.legend(['Curva de Densidade', 'Histograma'])

# Exibe o gráfico
plt.show()
```

Aqui se inicia uma análise por categoria

```python
df["Category"].value_counts()
```

O gráfico busca ajudar a visualizar a distribuição das categorias de produtos no conjunto de dados e a responder a perguntas como: Qual categoria é a mais comum? Como as categorias estão distribuídas?

```python
# Cria uma figura com o tamanho especificado
plt.figure(figsize=(20, 6))

# Cria um gráfico de barras que mostra a contagem de categorias da coluna 'Category' no DataFrame 'df'
ax = df["Category"].value_counts().plot(kind = 'bar', color = cores, rot = 0)

# Define os rótulos do eixo x como as categorias desejadas
ax.set_xticklabels(('Roupas', 'Acessórios', 'Calçados', 'Roupas Externas'))

# Adiciona rótulos com o valor de cada barra acima das barras
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha='center', va='bottom', color='black')

# Define o tamanho das fontes dos rótulos dos eixos
ax.tick_params(axis='both', labelsize=15)

# Define o rótulo do eixo x
plt.xlabel('Categoria', weight="bold", color="black", fontsize=14, labelpad=20)

# Define o rótulo do eixo y
plt.ylabel('Número de Ocorrências', weight="bold", color="black", fontsize=14, labelpad=20)

# Exibe o gráfico
plt.show()
```

```python
df["Item Purchased"].value_counts()
```

O gráfico vai ajudar a visualizar quais itens foram comprados com mais frequência no conjunto de dados e a entender a distribuição das compras em diferentes itens.

```python
# Cria uma figura com o tamanho especificado
plt.figure(figsize=(16, 7))

# Cria um gráfico de barras horizontais que mostra a contagem dos itens comprados
df["Item Purchased"].value_counts().sort_values(ascending=True).plot(kind='barh', color=sns.color_palette('tab20'), edgecolor='black')

# Define o rótulo do eixo y como "Item Purchased"
plt.ylabel('Item Comprado', fontsize=16)

# Define o rótulo do eixo x como "Número de Ocorrências"
plt.xlabel('\nNúmero de Ocorrências', fontsize=16)

# Define o título do gráfico
plt.title('Itens Comprados\n', fontsize=16)

# Define o tamanho e rotação dos rótulos do eixo x
plt.xticks(rotation=0, ha='center', fontsize=16)

# Ajusta o layout para evitar cortes de texto
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

quais cores são mais comuns nos itens comprados no conjunto de dados?

```python
# Cria uma figura com o tamanho especificado
plt.figure(figsize=(16, 6))

# Cria um gráfico de barras horizontais que mostra a contagem das cores mais frequentes (as 10 principais)
df["Color"].value_counts()[:10].sort_values(ascending=True).plot(kind='barh', color=sns.color_palette('tab20'), edgecolor='black')

# Define o rótulo do eixo x como "Cor"
plt.xlabel('Cor', weight="bold", color="black", fontsize=14, labelpad=20)

# Define o rótulo do eixo y como "Número de Ocorrências"
plt.ylabel('\nNúmero de Ocorrências', weight="bold", color="black", fontsize=14, labelpad=20)

# Define o tamanho e rotação dos rótulos do eixo x
plt.xticks(rotation=0, ha='center', fontsize=16)

# Ajusta o layout para evitar cortes de texto
plt.tight_layout()

# Exibe o gráfico
plt.show()
```
