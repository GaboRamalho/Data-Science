```python
dados = df2[['Age','Gender', 'JobPref', 'JobRoleInterest',
               'EmploymentField', 'JobRelocateYesNo',
               'HoursLearning', 'MoneyForLearning']]
dados
```

```python
# Create a histogram of 'Age' with 20 bins and light blue bars
plt.hist(dados['Age'], bins=20, color = 'lightblue', alpha = 0.5, edgecolor = 'blue')

# Label the Y-axis
plt.ylabel('Frequency')


# Set the title of the histogram
plt.title('Age distribution')

# Add a legend to explain the meaning of alpha
plt.legend(['Relationship between age distribution and frequency'], loc = 'best', fontsize = 12)

# Display the histogram
plt.show()
```

```python
# Import the necessary library
import matplotlib.pyplot as plt

# Create a scatter plot to visualize the relationship between 'HoursLearning' and 'MoneyForLearning'
plt.scatter(dados['HoursLearning'], dados['MoneyForLearning'], 
            color='green', alpha=0.5)

# Label the X and Y axes and add a title
plt.xlabel('Hours of Learning')
plt.ylabel('Money Spent for Learning')
plt.title('Relationship between Hours of Learning and Money Spent')

# Add a dashed grid to the Y-axis for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the scatter plot
plt.show()
```

```python
colors  = ['black', 'dimgray','dimgrey', 'gray', 'grey', 
         'darkgray', 'darkgrey', 'silver', 
          'lightgray', 'lightgrey']
```

```python
# Set the figure size to (15, 8)
plt.figure(figsize=(15, 8))

# Create a bar plot of 'JobPref' value counts with light blue bars and blue edges
dados['JobPref'].value_counts().plot(kind='bar', color = colors)

# Label the X and Y axes
plt.xlabel('Job Preference')
plt.ylabel('Count')

# Set the title of the plot
plt.title('Job Preference')

# Rotate the X-axis labels for better readability
plt.xticks(rotation=45)

# Display the bar plot
plt.show()
```
