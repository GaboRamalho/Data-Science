```python
import matplotlib.pyplot as plt
```

```python
#stack plots!

dias = [1, 2, 3, 4, 5]
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5, 3]
trabalhar = [7,8, 7, 2, 2]
passear = [8, 5, 7, 8, 13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear,
              labels = ['dias', 'dormir','comer', 'trabalhar', 'passear'],
              colors = ['m', 'c', 'r', 'k', 'b']) #each letter represents one color
plt.legend(loc = 'upper left')
plt.show()
```

```python
#Pie Chart 

porcentagens = [70, 20, 20, 130]
cidades = ['São Paulo', 'Recife', 'Rio de Janeiro', 'Salvador']
colunas = ['c', 'm', 'r', 'k'] #color used in plot
```

```python
plt.pie(porcentagens, labels = cidades, colors = colunas, startangle = 90, shadow = True, explode = (0, 0.15, 0, 0))
#explode mark one of the slice for when you wanna talk about it.
plt.show()
```

```python
import numpy as np
```

```python
# Plot and Scatter

fig = plt.figure()

ax1 = fig.add_subplot(1,2,1) #plot
ax1.plot(np.random.randn(50), color = 'red')

ax2 = fig.add_subplot(1,2,2) #scatterplot
ax2.scatter(np.arange(50), np.random.randn(50))

plt.show()

```

```python
# axle control 

x = np.linspace(0, 1., 100)
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("axis with standard range")

x = np.linspace(-1, 1., 100)
axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 1])
axes[1].set_xlim([-1, 1])
axes[1].set_title('Custom Axle')
```
