```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent', "Mcdonald's",
             'Visa', 'Facebook', 'Alibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062,819.573, 705.646,
            611.460, 213.023,
            196.526, 196.526, 186.421,
            169.966, 124.273]
```

```python
serie5 = pd.Series(bilhoes, index = empresas)
```

```python
serie5
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhoes')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhoes')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhoes')
plt.show()
```

```python
#mudando a cor e legendado

plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = '25')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('bilhoes', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = '13')
plt.show()
```

```python
#chaging color again

plt.plot(empresas, bilhoes, color = 'green', lw = 4, marker = 'o') #lw for chaging the line 
plt.title('Empresas mais ricas do mundo', fontsize = '25')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('bilhoes', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = '13')
plt.show()
```

```python
plt.plot(empresas, bilhoes, color = 'green', lw = 4, marker = 'o') #lw for chaging the line 
plt.title('Empresas mais ricas do mundo', fontsize = '25')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('bilhoes', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = '13')
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes, color = 'green')  
plt.title('Empresas mais ricas do mundo', fontsize = '25')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('bilhoes', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'best', fontsize = '13')
plt.show()
```

```python
cores = ['black', 'dimgray','dimgrey', 'gray', 'grey', 
         'darkgray', 'darkgrey', 'silver', 
          'lightgray', 'lightgrey']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')  
plt.title('Empresas mais ricas do mundo', fontsize = '25')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('bilhoes', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'best', fontsize = '13')
plt.show()
```
