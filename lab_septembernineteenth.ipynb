Which companies pollute the most?
https://brandaudit.breakfreefromplastic.org/brand-audit-2020/

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#variable to input the companies listed in the pdf

companies = pd.Series(['Coca-Cola', 'Pepsico', 'Nestle',
             'Unilever', 'Mondelez', 'Mars',
             'P&G', 'Phillip Morris', 'Colgate-Palmolive',
             'Perfetti van Melle'])
```

```python
#variable to input Numbers of countries in which waste was found

number_of_countries = pd.Series([51, 43, 37,
                                37, 34, 32,
                                29, 28, 24,
                                24])
```

```python
#variable to input pieces of waste recorded

pieces_of_waste = pd.Series([13834, 5155, 8633,
                            5558, 1171, 687,
                            3535, 2593, 5991,
                            465])
```

```python
# Create a DataFrame to combine the Series
df = pd.DataFrame({
    'Company': companies,
    'Number_of_Countries': number_of_countries,
    'Pieces_of_Waste': pieces_of_waste
})
```

```python
df
```

```python
colors  = ['black', 'dimgray','dimgrey', 'gray', 'grey', 
         'darkgray', 'darkgrey', 'silver', 
          'lightgray', 'lightgrey']
```

```python
plt.rcParams['figure.figsize'] = (15, 8)  # Set the figure size

# Create the bar plot with 
bars = plt.bar(df['Company'], df['Number_of_Countries'], color = colors, align='center')

# Set Y-axis limit to 0 to 60
plt.ylim(0, 60)

# Set labels and title
plt.ylabel('Country', fontsize=20)
plt.xlabel('Companies', fontsize=20)
plt.title('Who pollute the most?', fontsize=25)
plt.legend(['Number of countries affected per company'], loc='best', fontsize='13')

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 8)  # Set the figure size

# Create the bar plot with 
bars = plt.bar(df['Company'], df['Pieces_of_Waste'], color = colors, align='center')

# Set X-axis limit from 0 to 14000
plt.ylim(0, 14000)

# Set labels and title
plt.ylabel('Plastic', fontsize=20)
plt.xlabel('Companies', fontsize=20)
plt.title('Who pollute the most?', fontsize=25)
plt.legend(['Quantity of plastic per company'], loc='best', fontsize='13')

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

This code enables us to gauge the global influence of various companies, as outlined in "BRANDED Volume III: Demanding Corporate Accountability for Plastic Pollution". 
By analyzing the data, it becomes evident that Coca-Cola emerge as the company with the most significant impact, having nearly 14.000 plastic items being removed from ocean across 50 different countries.
